```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python


# Read CSV file

cumulative_distance = pd.read_csv('./csv/cumulativeDistanceTravel_baseline0.csv')
locational_cost = pd.read_csv('./csv/locationalCost_baseline0.csv')
health_cost = pd.read_csv('./csv/healthCost_baseline0.csv')
mobility_cost = pd.read_csv('./csv/mobilityCost_baseline0.csv')
range_cost = pd.read_csv('./csv/rangeCost_baseline0.csv')
proposed_cost = pd.read_csv('./csv/proposedCost_baseline0.csv')

# Display first 5 rows
df_cost = pd.concat([cumulative_distance,locational_cost, health_cost, mobility_cost,range_cost, proposed_cost], axis=1)
df_cost.columns=['cumulative_distance','locational_cost', 'health_cost', 'mobility_cost','range_cost', 'proposed_cost']
print(df_cost.head())

```

```python


plt.plot(df_cost.index, df_cost.locational_cost, label ="a")
plt.plot(df_cost.index, df_cost.health_cost, label ="b")

# Add title and axis labels
plt.title("Locational Cost (scenerio 1)", fontsize=14)   # Set title
plt.xlabel("Iteration Number", fontsize=12) # X-axis label
plt.ylabel("Hg", fontsize=12)              # Y-axis label
# Add a legend
plt.legend()

# Show the grid for better readability
plt.grid(True)
plt.savefig("./plot/locational_cost_scenario1.png")
plt.show()
```
