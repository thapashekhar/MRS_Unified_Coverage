```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python


# Read CSV file

cumulative_distance = pd.read_csv( './csv/s6/cumulativeDistanceTravel.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
locational_cost = pd.read_csv( './csv/s6/locationalCost.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
health_cost = pd.read_csv('./csv/s6/healthCost.csv',header=None, converters={0: lambda x: float(x.strip("[]"))})
mobility_cost = pd.read_csv( './csv/s6/mobilityCost.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
range_cost = pd.read_csv('./csv/s6/rangeCost.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
proposed_cost = pd.read_csv('./csv/s6/proposedCost.csv', header=None,converters={0: lambda x: float(x.strip("[]"))})

# Display first 5 rows
df_cost = pd.concat([cumulative_distance,locational_cost, health_cost, mobility_cost,range_cost, proposed_cost], axis=1)
df_cost.columns=['cumulative_distance','locational_cost', 'health_cost', 'mobility_cost','range_cost', 'proposed_cost']
print(df_cost.head())
```

```python
cumulative_distance1 = pd.read_csv( './csv/s6/cumulativeDistanceTravelB0.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
locational_cost1 = pd.read_csv( './csv/s6/locationalCostB0.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
health_cost1 = pd.read_csv('./csv/s6/healthCostB0.csv',header=None, converters={0: lambda x: float(x.strip("[]"))})
mobility_cost1 = pd.read_csv( './csv/s6/mobilityCostB0.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
range_cost1 = pd.read_csv('./csv/s6/rangeCostB0.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
proposed_cost1 = pd.read_csv('./csv/s6/proposedCostB0.csv', header=None,converters={0: lambda x: float(x.strip("[]"))})

# Display first 5 rows
df_cost1 = pd.concat([cumulative_distance1,locational_cost1, health_cost1, mobility_cost1,range_cost1, proposed_cost1], axis=1)
df_cost1.columns=['cumulative_distance','locational_cost', 'health_cost', 'mobility_cost','range_cost', 'proposed_cost']
print(df_cost1.head())
```

```python
cumulative_distance11 = pd.read_csv( './csv/s6/cumulativeDistanceTravelB3.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
locational_cost11 = pd.read_csv( './csv/s6/locationalCostB3.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
health_cost11 = pd.read_csv('./csv/s6/healthCostB3.csv',header=None, converters={0: lambda x: float(x.strip("[]"))})
mobility_cost11 = pd.read_csv( './csv/s6/mobilityCostB3.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
range_cost11 = pd.read_csv('./csv/s6/rangeCostB3.csv', header=None, converters={0: lambda x: float(x.strip("[]"))})
proposed_cost11 = pd.read_csv('./csv/s6/proposedCostB3.csv', header=None,converters={0: lambda x: float(x.strip("[]"))})

# Display first 5 rows
df_cost11 = pd.concat([cumulative_distance11,locational_cost11, health_cost11, mobility_cost11,range_cost11, proposed_cost11], axis=1)
df_cost11.columns=['cumulative_distance','locational_cost', 'health_cost', 'mobility_cost','range_cost', 'proposed_cost']
print(df_cost11.head())
```

```python


plt.plot(df_cost.index, df_cost.proposed_cost, label ="Peoposed" )
plt.plot(df_cost1.index, df_cost1.proposed_cost, label ="Baseline0")
plt.plot(df_cost11.index, df_cost11.proposed_cost, label ="Baseline3")

# Add title and axis labels
plt.title("Proposed Cost (scenerio 6)", fontsize=14)   # Set title
plt.xlabel("Iteration Number", fontsize=12) # X-axis label
plt.ylabel("Proposed Cost", fontsize=12)              # Y-axis label
# Add a legend
plt.legend()

# Show the grid for better readability
plt.grid(True)
plt.savefig("./plot/locational_cost_scenario6.png")
plt.show()
```
