```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.gridspec as gridspec
import matplotlib.image as mpimg  # To read images
```

```python

# Define image paths (Replace with your file paths)
image_paths = ["./plot/plot/s1.png","./plot/plot/baseline0_s1.png", "./plot/plot/baseline1_s1.png", "./plot/plot/baseline2_s1.png", "./plot/plot/baseline3_s1.png", "./plot/plot/baseline4_s1.png"]

# Read images using matplotlib
images = [mpimg.imread(img) for img in image_paths]

# Create a figure
fig = plt.figure(figsize=(10, 8))

# Define GridSpec layout (2 rows, 3 columns)
gs = gridspec.GridSpec(3, 3, figure=fig)

# Assign images to subplots
ax1 = fig.add_subplot(gs[0, 0])
ax1.imshow(images[0])
#ax1.set_title("Image 1")
ax1.axis("off")

ax2 = fig.add_subplot(gs[0, 1])
ax2.imshow(images[1])
#ax2.set_title("Image 2")
ax2.axis("off")

ax3 = fig.add_subplot(gs[1, 0])
ax3.imshow(images[2])
#ax3.set_title("Image 3")
ax3.axis("off")

ax4 = fig.add_subplot(gs[1, 1])
ax4.imshow(images[3])
#ax4.set_title("Image 4")
ax4.axis("off")

ax5 = fig.add_subplot(gs[2, 0])  # Spanning two columns
ax5.imshow(images[4])
#ax5.set_title("Image 5 (Spanning Two Columns)")
ax5.axis("off")

ax5 = fig.add_subplot(gs[2, 1])  # Spanning two columns
ax5.imshow(images[5])
#ax5.set_title("Image 5 (Spanning Two Columns)")
ax5.axis("off")

fig.suptitle("Scenario Number = 1", fontsize=14, fontweight='bold')

# Adjust layout for better spacing
plt.tight_layout()
plt.savefig("./plot/plot/output_figure.png", dpi=300, bbox_inches="tight")
plt.show()

```
